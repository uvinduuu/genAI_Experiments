```python
#splitting using a for loop

text = "text here"

chunks = []

s=""
for word in text:
    s += word + " "
    if len(s)>200:
        chunks.append(s)
        s = ""

chunks.append(s)

#this method is also not recommended. Therefore we are using an inbuilt method in langchain
```

```python
from langchain.text_splitter import CharacterTextSplitter

splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size=200,
    chunk_overlap=0
)

chunks = splitter.split_text(text)
#len(chunks) #number of chunks
    
```

This method is also not recommmended as the separator is not much effective. If the text doesn't use '\n' in less than 200 words, the chunk exceeds it limit. Therefore we are using the new recursive method.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

r_splitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n", "\n", " "], #multiple separators
    chunk_size=200,
    chunk_overlap=0
)    

chunks = r_splitter.split_text(text)
```

Here first it splits with the first separator and like wise it splits with all the separators as the chunk size get below 200. Because of that here we are getting chunks with the size very less than 200(Ex:10). That happens because in the splitter first it splits into 420,390,... . Then it splits again 230, 190, ... . In the next separator round it splits that 198, 32, ... .

```python
#converting text in the csv file to a vector database

from sentence_transformers import SentenceTransformer

encoder = SentenceTransformer("all-mpnet-base-v2")
```

```python
vectors = encoder.encode(df.text)
vectors(shape)
```

```python
dim = vectors.shape[1] #size of a one vector, vector.shapr[0] refers to number of vectors
```

```python
import faiss

index = faiss.IndexFlat2(dim) #getting the Euclidean distance (It is a measure of the straight-line distance between two points in Euclidean space; L2 distance)

#this indexing allows us to do fast searching between vectors
```

```python
search_query = "I want to buy a trouser"

vec = encoder.encode(search_query)
vec.shape
```

```python
#as the above one is an one dimensional array, we have to convert it into a two dimensional array.
import numpy as np
svec = np.array(vec).reshape(1,-1)
svec.shape
                        
```
